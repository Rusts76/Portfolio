# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
#подключаем необходимые библиотеки
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/datasets/real_estate_data.csv', '\t')
#получаем общую информацию о базе данных и отображаем первые пять строк
data.info()
display(data.head())
#проверяем базу данных на явные дубликаты
print('Количество дубликатов', data.duplicated().sum())
```

### Вывод

last_price - нормальная информация

total_area - нормальная информация

living_area - нормальная информация

is_apartment - скорее всего остальные записи - не апартаменты, можно заполнить False

balcony - информация о количестве балконов, больше всего предложений имеет один балкон, есть записи с количеством балконов 4 и 5, что скорее всего для квартир с количеством комнат менее 4 и 5 является ошибкой.

days_exposition - максимальное значение 1174 дней, минимальное 1.

## Предобработка данных

```python
#методом to_datetime() переводит строки в даты в столбце data['first_day_exposition']
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d %H:%M')
#заполнение пропусков в столбце data['balcony'] значением 0
data['balcony'] = data['balcony'].fillna('0')
#заполнение пропусков в столбце data['is_apartment'] значением False
data['is_apartment'] = data['is_apartment'].fillna('False')
```

### Вывод
Преобразован формат столбца 'first_day_exposition', получено корректное значение даты, которое можно использовать для дальнейшего анализа.

Пустые значения в столбце 'balcony' заполнены 0.

Пустые значения в столбце 'is_apartment' заполнены False.

## Расчёты и добавление результатов в таблицу

```python
#добавляем цену квадратного метра
data['last_price_m2'] = data['last_price'] / data['total_area']
data['last_price_m2'] = data['last_price_m2'].astype('int')
#добавляем день недели
data['weekday'] = data['first_day_exposition'].dt.weekday
c
data['month'] = data['first_day_exposition'].dt.month
#добавляем год
data['year'] = data['first_day_exposition'].dt.year
#добавляем отношение жилой площади к общей площади
data['living_total'] = data['living_area'] / data['total_area']
#добавляем отношение площади кухни к общей площади
data['kitchen_total'] = data['kitchen_area'] / data['total_area']
#создаем функцию, которая проводит категоризацию этажа
def floor_category(floor):
    if floor == 1:
        return 'первый'
    else:
        return 'другой'
#применение функции методом apply() к данныем из столбца data['floor'] и внесение итогового результата в столбец
#data['floor_category']
data['floor_category'] = data['floor'].apply(floor_category)
#создаем вспомогательный столбец для выявления последних этажей, он будет образован путем деления data['floor'] 
#на data['floors_total'], когда это значение будет равно единице, значит этаж - последний
data['floor_last'] = data['floor'] / data['floors_total']
#выбираем строки, в которых значение data['floor_last'] равно 1
rows = data['floor_last'] == 1
#присваиваем значение последний для этих строк в столбце data['floor_category']
data.loc[rows, 'floor_category'] = 'последний'
print(data['floor_category'].value_counts())
```

### Вывод
Добавлены дополнительные столбцы для анализа:
- цена квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.
